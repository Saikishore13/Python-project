## 1B: Library Book Management System

```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
```

```python
def Database():
    global conn, cursor
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS lbms (Author_name TEXT,booktitle TEXT,year INTEGER,ISBN INTEGER)")

def display_record():
    present.delete(*present.get_children())
    Database()
    cursor.execute('select * from lbms order by booktitle ASC')
    fetch = cursor.fetchall()
    for data in fetch:
        present.insert('',"end",values=(data[0],data[1],data[2],data[3]))
    conn.close()
    l1.config(text='please see the book',fg='green')
                   
def add_new_book():
    Database()
    if Author_name.get(1.0,"end") != '' and booktitle.get(1.0,"end") != '' and year.get(1.0,"end") != '' and ISBN.get(1.0,"end") != '' :
        cursor.execute('insert into lbms values(?,?,?,?)',(Author_name.get(1.0,"end"),booktitle.get(1.0,"end"),int(year.get(1.0,"end")),int(ISBN.get(1.0,"end"))))
        conn.commit()
        delete()
        conn.close()
        l1.config(text='Book added successfully',fg='green')
    else:
        l1.config(text='please enter all the values to add book',fg='red')
        delete()

def search_book():
    present.delete(* present.get_children())
    Database() 
    if str(Author_name.get(1.0,"end"))!="":
        name = " ' "+str(Author_name.get(1.0,"end")).title()+" ' "
        temp ="select * from lbms WHERE Author_name"+name
        cursor.execute(temp)
        fetch = cursor.fetchall() 
        for data in fetch:
            present.insert('',"end",values=(data[0],data[1],data[2],data[3])) 
        l1.config(text='please see the book',fg='green')
    elif str(booktitle.get(1.0,"end"))!="":
        name = " ' "+str(booktitle.get(1.0,"end")).title()+" ' "
        temp = "select * from lbms WHERE booktitle"+name
        cursor.execute(temp)       
        fetch = cursor.fetchall()
        for data in fetch:
            present.insert('',"end",values=(data[0],data[1],data[2],data[3]))
    elif str(year.get(1.0,"end"))!="":
        name = " ' "+str(year.get(1.0,"end")).title()+" ' "
        temp = "select * from lbms WHERE year"+name
        cursor.execute(temp)       
        fetch = cursor.fetchall()
        for data in fetch:
            present.insert('',"end",values=(data[0],data[1],data[2],data[3]))
    elif str(ISBN.get(1.0,"end"))!="":
        name = " ' "+str(ISBN.get(1.0,"end")).title()+" ' "
        temp = "select * from lbms WHERE ISBN"+name
        cursor.execute(temp)       
        fetch = cursor.fetchall()
        for data in fetch:
            present.insert('',"end",values=(data[0],data[1],data[2],data[3]))                           
    else:
        l1.config(text='please enter atleast single identity',fg='red')
    delete()                      
    conn.close()                             
               
def issue_book():
    Database()
    selected_item=present.selection()                     
    if len(selected_item)>0:
        present.delete(selected_item)
        l1.config(text='selected book is issued',fg='green')
    else:
        l1.config(text='please select book from library',fg='red')
    delete()                        
    conn.close()                    
def delete_book():
    Database()                             
    if Author_name.get(1.0,"end") =='' or booktitle_name.get(1.0,"end") =='' or year.get(1.0,"end") =='' or ISBN.get(1.0,"end") =='':
        l1.config(text='please enter all entries',fg='red')
        delete()  
    else:
        Author ='"' +str(Author_name.get(1.0,"end"))+'"'                           
        Book ='"' +str(booktitle.get(1.0,"end"))+'"'            
        y =str(year.get(1.0,"end"))             
        I =str(ISBN.get(1.0,"end"))
        temp = 'delete from lbms Author_name like '+Author+'and booktitle like'+  Book+'and year'+y+'and ISBN '+I
                                  
        cursor.execute(temp)
        conn.commit()                            
        l1.config(text='Books deleted successfully',fg='red')
        delete()                   
    conn.close()      
                   
def delete():
    Author_name.delete(1.0,"end")                       
    booktitle.delete(1.0,"end")              
    year.delete(1.0,"end")              
    ISBN.delete(1.0,"end") 
                           
def exit():
    result = tkMessageBox.askquestion('Do you want to exit? (y/n)',icon='warning')                      
    if result == 'yes':
        root.destroy()                    
        exit()            

  
```
